```python
import os

from datetime import datetime

from elasticsearch.helpers import parallel_bulk, streaming_bulk
import elasticsearch
import xmltodict
```

```python
PATH = '/Users/aman/workspace/DataMining/HW5/DUC2001'
from pyexpat import ExpatError
es = elasticsearch.Elasticsearch(http_auth=('elastic', 'elastic'))

def get_formated_data(file_name):
    structured_json_body = {
            '_op_type': 'index',
            '_index': 'duc2001',
            '_type': 'document',
            '_id': file_name,
    }
    with open(PATH+'/'+file_name) as fd:
        
        try:
            doc = xmltodict.parse(fd.read())
            doc = doc['DOC']
            json= dict()
            json['doc_id'] = doc['DOCNO']
            json['doc_text'] = str(doc['TEXT'])
            with open(PATH+'/Summaries/'+file_name.lower()+'.txt') as fd_sum:
                json['gold_summary']=fd_sum.read()
            structured_json_body.update(json)
        except ExpatError:
            print file_name
        except KeyError:
            print file_name,fd.read()
    return structured_json_body
doc_list = []
for (dir_path, dir_names, file_names) in os.walk(PATH):
    if dir_path.split('/')[-1] == 'DUC2001':
        for doc in file_names:
            doc_list.append(get_formated_data(doc))
for (res, i) in streaming_bulk(client=es, raise_on_exception=True, raise_on_error=True,
                              actions=doc_list):
    print(res, i)
```
